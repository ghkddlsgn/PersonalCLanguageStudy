```python
#handling missing values
print(df.isnull().any())
print(df.isnull().any(axis=1))
print(df.isnull().sum())
```

```python
#replace null value to 0
df.fillna(0)
```

```python
#object basically mean string
df.dtypes
```

```python
#remain col
df = df.rename(columns={'Date':'Sales Date'})
df.head()
```

```python
#change datatypes
df['Quantity2']=df['Quantity'].astype(float)
df.head()
```

```python
df['Quantity3']=df['Quantity'].apply(lambda x:x**2)
df.head(5)
```

```python
#Data aggregating and grouping
grouped_mean = df.groupby('Product_Name')['Unit_Price'].mean()
print(grouped_mean)
```

```python
newgroup = df.groupby(["Region", 'Product_Name']).agg({
    'Unit_Price': 'first',  # since it's same within group
    'Total_Amount': 'sum'
})
newgroup['total_sale'] = newgroup['Unit_Price'] * newgroup['Total_Amount']
print(newgroup)
```
